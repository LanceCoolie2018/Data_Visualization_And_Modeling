```python
import pandas as pd

songs_df = pd.read_csv('spotify-2023.csv', encoding='ISO-8859-1')

songs_df['bpm'].plot(kind= 'hist',
                    title = 'BPM',
                       ylabel='Number of songs',
                       xlabel='BPM Range',
                            edgecolor='black',
                           figsize=(10, 6))
```

```python
The best range for beats per minute is between 120 and 140 BPM so I believe that 130 will yield the best results.
```

```python
songs_df['key'].value_counts().plot(kind= 'bar',
                    title = 'Key')
```

```python
From the chart, C sharp is used the most in songs.
```

```python
songs_df['mode'].value_counts().plot(kind= 'bar',
                    title = 'Mode')
```

```python
Major is used more than 500 times where minor is barely reaching 400.
```

```python
songs_df['danceability_%'].plot(kind= 'hist',
                    title = 'Danceability %',
                            edgecolor='black',
                           figsize=(10, 6))
```

```python
Based on the data there are only a little over 50 songs that are 90% danceable to and most songs fall between 70% - 80%.
```

```python
songs_df['energy_%'].plot(kind= 'hist',
                    title = 'Energy',
                            edgecolor='black',
                           figsize=(10, 6))
```

```python
Looking at the energy graph you can see a very close relation between the danceability of a song and the energy it has.
```

```python
songs_df['speechiness_%'].plot(kind= 'hist',
                    title = 'Speechiness',
                               ylabel = 'Number of songs',
                               xlabel = '% of spoken words',
                            edgecolor='black',
                           figsize=(10, 6))
```

```python
Most songs are not very wordy and instead allow the music to do the talking.
```

```python
songs_df['acousticness_%'].plot(kind= 'hist',
                    title = 'Acoustics',
                               ylabel = 'Number of songs',
                               xlabel = '% of Acoustics',
                            edgecolor='black',
                           figsize=(10, 6))
```

```python
Most songs also like to have less Acoustical noise with very few making it the majority of the song.
```

```python
engineered_songs = songs_df.loc[ (songs_df['key'] == 'C#') &
                           (songs_df['bpm'] > 120) &
                           (songs_df['bpm'] < 140) &
                            (songs_df['mode'] == 'Major')&
                            (songs_df['danceability_%'] > 70)&
                            (songs_df['energy_%'] > 70)&
                            (songs_df['acousticness_%'] < 20)&
                            (songs_df['speechiness_%'] < 10)]

if len(engineered_songs) > 0:
    display(engineered_songs)
```
